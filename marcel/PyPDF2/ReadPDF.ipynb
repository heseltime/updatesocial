```python
import fitz  # PyMuPDF

def analyze_pdf(pdf_file_path):
    doc = fitz.open(pdf_file_path)

    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        tree = page.get_text("dict")

        # Extracting the text and font information
        blocks = tree["blocks"]
        for b in blocks:
            if b["type"] == 0:  # Text block
                for l in b["lines"]:
                    for s in l["spans"]:
                        print("Text:", s["text"])
                        print("Font:", s["font"])

    doc.close()

# Example usage
analyze_pdf(file)
```
